```python
# Import libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Dataset Paths
dataset_path = "dataset/train"  # Update this to your dataset's train folder path

# Classes (update these to match the exact folder names in your dataset)
classes = ['bacterial_leaf_blight', 'brown_spot', 'healthy', 'leaf_blast', 'sheath_blight']
num_classes = len(classes)

# Image Preprocessing
def preprocess_image(img_path):
    try:
        # Load the image with reduced memory consumption
        img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        if img is None:
            raise ValueError(f"Image at {img_path} could not be read. Check the file.")

        # Resize the image to 128x128 pixels
        img_resized = cv2.resize(img, (128, 128))

        # Normalize pixel values to [0, 1]
        img_normalized = img_resized / 255.0
        return img_normalized
    except Exception as e:
        print(f"Error processing image {img_path}: {e}")
        return None  # Skip this image

# Load dataset
def load_dataset():
    images = []
    labels = []
    for class_label, class_name in enumerate(classes):
        class_path = os.path.join(dataset_path, class_name)
        if not os.path.exists(class_path):
            raise FileNotFoundError(f"Class folder not found: {class_path}")
        for img_file in os.listdir(class_path):
            img_path = os.path.join(class_path, img_file)
            if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):  # Ensure valid image files
                processed_img = preprocess_image(img_path)
                if processed_img is not None:  # Skip if preprocessing failed
                    images.append(processed_img)
                    labels.append(class_label)
    return np.array(images), np.array(labels)

# Load the dataset
print("Loading dataset...")
try:
    images, labels = load_dataset()
    print(f"Dataset loaded. Total images: {len(images)}")
except Exception as e:
    print(f"Error loading dataset: {e}")
    raise

# Convert labels to one-hot encoding
labels_one_hot = to_categorical(labels, num_classes)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels_one_hot, test_size=0.2, random_state=42)

# Sequential Model Architecture
model = Sequential()

# Add convolutional layers with MaxPooling
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# Flatten the output and add Dense layers
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Regularization to reduce overfitting
model.add(Dense(num_classes, activation='softmax'))  # Output layer

# Compile the Model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Model Summary
model.summary()

# Train the Model
batch_size = 32
epochs = 20
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test))

# Evaluate the Model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {accuracy * 100:.2f}%")

# Plot Training History
plt.figure(figsize=(12, 4))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Save the Model
model.save("paddy_leaf_disease_sequential_model.h5")
print("Model saved as 'paddy_leaf_disease_sequential_model.h5'")

```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Confusion Matrix
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

conf_matrix = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=classes)
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
plt.title("Confusion Matrix")
plt.show()
```

```python
y_pred_labels = np.argmax(y_pred, axis=1)  # Convert probabilities to class indices
y_true_labels = np.argmax(y_test, axis=1)  # Convert one-hot encoded labels to class indices
```

```python
from sklearn.metrics import classification_report

# Generate the classification report
report = classification_report(y_true_labels, y_pred_labels, target_names=classes)
print(report)
```

```python
import os
import cv2
import numpy as np
import pickle
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from fpdf import FPDF

# Path to the saved model
model_path = "paddy_leaf_disease_sequential_model.h5"  # Update this if your model is named differently
try:
    model = load_model(model_path)
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")
    exit()

# Path to the pickle file with treatment suggestions
treatment_file_path = "/content/Disease_Classes_and_Treatment_Suggestions.pkl"  # Update with the correct path

# Load treatment suggestions
try:
    with open(treatment_file_path, "rb") as file:
        treatment_data = pickle.load(file)
    print("Treatment data loaded successfully.")
    print("Available Treatment Classes:", list(treatment_data.keys()))
except Exception as e:
    print(f"Error loading treatment data: {e}")
    treatment_data = {}  # Default to empty if loading fails

# Classes (same as in the trained model)
classes = ['bacterial_leaf_blight', 'brown_spot', 'healthy', 'leaf_blast', 'sheath_blight']

# Image Dimensions
image_size = (128, 128)

# Function to preprocess an image for testing
def preprocess_image_for_prediction(img_path):
    img = cv2.imread(img_path)
    if img is None:
        raise ValueError(f"Image at {img_path} could not be read. Check the file.")
    img_resized = cv2.resize(img, image_size)  # Resize to match training dimensions
    img_normalized = img_resized / 255.0      # Normalize pixel values to [0, 1]
    img_reshaped = np.expand_dims(img_normalized, axis=0)  # Add batch dimension
    return img_reshaped

# Function to display treatment suggestions
def display_treatment_suggestions(predicted_class):
    normalized_class = predicted_class.lower()
    treatment_data_normalized = {key.lower(): value for key, value in treatment_data.items()}

    if normalized_class in treatment_data_normalized:
        return treatment_data_normalized[normalized_class]
    else:
        print("\nNo treatment suggestions found for this class.")
        return None

# Function to generate a PDF report
def generate_pdf_report(image_path, predicted_class, predictions, suggestions, pdf_path):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    # Set Title
    pdf.set_font("Arial", style="B", size=16)
    pdf.cell(0, 10, "Prediction Report", ln=True, align="C")
    pdf.ln(10)  # Add a line space below the title

    # Add Image
    if os.path.exists(image_path):
        pdf.set_font("Arial", size=12)
        pdf.cell(0, 10, "Predicted Image:", ln=True, align="L")
        pdf.image(image_path, x=60, y=30, w=90)  # Center image on the page
        pdf.ln(90)  # Add space below the image

    # Add Prediction Result
    pdf.set_font("Arial", style="B", size=12)
    pdf.cell(0, 10, f"Predicted: {predicted_class} ({predictions[0][classes.index(predicted_class)] * 100:.2f}%)", ln=True, align="L")
    pdf.ln(10)  # Add a line space

    # Add Treatment Suggestions
    if suggestions:
        pdf.set_font("Arial", style="B", size=12)
        pdf.cell(0, 10, "--- Treatment Suggestions ---", ln=True, align="L")
        pdf.set_font("Arial", size=12)
        pdf.cell(0, 8, f"Disease Class: {predicted_class}", ln=True, align="L")
        pdf.cell(0, 8, f"Name: {suggestions.get('Name', 'N/A')}", ln=True, align="L")
        pdf.cell(0, 8, f"Symptoms: {suggestions.get('Symptoms', 'N/A')}", ln=True, align="L")
        pdf.cell(0, 8, f"Treatment (Brand Names): {suggestions.get('Treatment (Brand Names)', 'N/A')}", ln=True, align="L")
        pdf.cell(0, 8, f"How to Use: {suggestions.get('How to Use', 'N/A')}", ln=True, align="L")
        pdf.cell(0, 8, f"Recommendations: {suggestions.get('Recommendations', 'N/A')}", ln=True, align="L")
    else:
        pdf.set_font("Arial", size=12)
        pdf.cell(0, 10, "No treatment suggestions found.", ln=True, align="L")

    # Save the PDF
    pdf.output(pdf_path)
    print(f"PDF report saved at {pdf_path}")

# Function to make predictions
def predict_image(img_path):
    try:
        # Check if the image path exists
        if not os.path.exists(img_path):
            raise FileNotFoundError(f"Image not found at path: {img_path}")

        # Preprocess the image
        img = preprocess_image_for_prediction(img_path)

        # Predict using the trained model
        predictions = model.predict(img)
        predicted_class_index = np.argmax(predictions)
        predicted_class = classes[predicted_class_index]

        # Display the image and prediction
        img_display = cv2.imread(img_path)
        img_display = cv2.cvtColor(img_display, cv2.COLOR_BGR2RGB)  # Convert to RGB for display
        plt.imshow(img_display)
        plt.title(f"Predicted: {predicted_class} ({predictions[0][predicted_class_index] * 100:.2f}%)")
        plt.axis('off')
        plt.show()

        # Display treatment suggestions
        suggestions = display_treatment_suggestions(predicted_class)

        # Generate PDF
        pdf_path = f"Prediction_Report_{predicted_class}.pdf"
        generate_pdf_report(img_path, predicted_class, predictions, suggestions, pdf_path)

        return predicted_class, predictions
    except Exception as e:
        print(f"Error predicting image {img_path}: {e}")
        return None

# Test a single image
image_path = "/content/brown_spot (4).jpg"  # Update with your test image path
predicted_class, predictions = predict_image(image_path)
if predicted_class:
    print(f"\nPredicted Class: {predicted_class}")
    print(f"Prediction Probabilities: {predictions}")
else:
    print("Prediction failed.")
```
